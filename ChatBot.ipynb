```python
import re
import random
import sys  # Provides access to interpreter-level functions (used here for clean program termination)

# =========================
# GLOBAL USER DATA
# =========================
# Stores the user's name during the current chat session.
# This allows the chatbot to personalize responses.
user_name = None

# =========================
# KNOWLEDGE BASE
# =========================

GENERAL_RESPONSES = {
    "greeting": [
        "Hello! 👋 How can I assist you today?",
        "Hi there! What would you like to learn?",
        "Welcome! I am ready to help you."
    ],
    "thanks": [
        # Responses triggered when the user expresses gratitude
        "You're welcome! 😊 Let me know if you need anything else.",
        "Happy to help! Feel free to ask more questions.",
        "No problem at all. Good luck with your studies!"
    ],
    "day": [
        "I'm functioning optimally, thank you. How has your day been?",
        "All systems are running well. How are you today?"
    ],
    "default": [
        "Could you please clarify your question?",
        "That’s an interesting query. Can you explain further?",
        "Let’s explore that further. Kindly provide more details."
    ],
    "bye": [
        "Goodbye! 👋 Wishing you academic success.",
        "Session ended. Best of luck with your studies.",
        "Thank you for using the chatbot. Goodbye!"
    ]
}

# Mathematics topics
MATH_TOPICS = {
    "algebra": "Algebra deals with symbols and variables used to represent numbers and quantities.",
    "calculus": "Calculus is the branch of mathematics concerned with rates of change and accumulation.",
    "linear programming": "Linear programming is a mathematical technique for optimizing a linear objective function under given constraints.",
    "matrix": "A matrix is a rectangular array of numbers arranged in rows and columns."
}

# Computer Science topics
CS_TOPICS = {
    "algorithm": "An algorithm is a finite set of step-by-step instructions used to solve a problem.",
    "data structure": "A data structure is a method of organizing and storing data for efficient access and modification.",
    "python": "Python is a high-level, interpreted programming language known for readability and versatility.",
    "operating system": "An operating system manages computer hardware and software resources."
}

# Cybersecurity topics
CYBERSECURITY_TOPICS = {
    "malware": "Malware refers to malicious software designed to disrupt, damage, or gain unauthorized access to systems.",
    "phishing": "Phishing is a social engineering attack used to obtain sensitive information by deception.",
    "firewall": "A firewall is a network security device that monitors and controls network traffic.",
    "encryption": "Encryption is the process of converting data into a secure format to prevent unauthorized access."
}

# Course code information (common in university environments)
COURSE_CODES = {
    "csc101": "CSC101 – Introduction to Computer Science. Covers basic computing concepts and programming fundamentals.",
    "mat101": "MAT101 – General Mathematics. Focuses on algebra, calculus, and basic mathematical techniques.",
    "cyb201": "CYB201 – Introduction to Cybersecurity. Covers threats, vulnerabilities, and security principles."
}

# =========================
# CORE CHATBOT FUNCTION
# =========================

def get_response(user_input):
    """
    Determines the chatbot's response based on the user's input.
    Regular expressions are used for pattern matching.
    """
    global user_name
    user_input = user_input.lower()

    # -------------------------
    # NAME CAPTURE PATTERN
    # -------------------------
    # Detects statements like "my name is John" and stores the name for personalization.
    name_match = re.search(r"my name is ([a-zA-Z]+)", user_input)
    if name_match:
        user_name = name_match.group(1).capitalize()
        return f"Nice to meet you, {user_name}! 😊 How can I assist you today?"

    # -------------------------
    # GREETING PATTERN
    # -------------------------
    if re.search(r"\b(hi|hello|hey)\b", user_input):
        if user_name:
            return f"Hello, {user_name}! 👋 What would you like to learn today?"
        return random.choice(GENERAL_RESPONSES["greeting"])

    # -------------------------
    # GRATITUDE / THANKS PATTERN
    # -------------------------
    # Responds politely when the user says "thanks" or "thank you".
    if re.search(r"\b(thanks|thank you|thx)\b", user_input):
        return random.choice(GENERAL_RESPONSES["thanks"])

    # -------------------------
    # DAY INQUIRY PATTERN
    # -------------------------
    if re.search(r"\b(how are you|how is your day)\b", user_input):
        if user_name:
            return f"I'm doing well, {user_name}. How has your day been?"
        return random.choice(GENERAL_RESPONSES["day"])

    # -------------------------
    # COURSE CODE PATTERN
    # -------------------------
    # Allows queries such as "What is CSC101?" or "Explain MAT101".
    for code in COURSE_CODES:
        if re.search(rf"\b{code}\b", user_input):
            return f"📚 Course Information:\n{COURSE_CODES[code]}"

    # -------------------------
    # MATH TOPICS
    # -------------------------
    for topic in MATH_TOPICS:
        if re.search(rf"\b{topic}\b", user_input):
            return f"📘 Mathematics:\n{MATH_TOPICS[topic]}"

    # -------------------------
    # COMPUTER SCIENCE TOPICS
    # -------------------------
    for topic in CS_TOPICS:
        if re.search(rf"\b{topic}\b", user_input):
            return f"💻 Computer Science:\n{CS_TOPICS[topic]}"

    # -------------------------
    # CYBERSECURITY TOPICS
    # -------------------------
    for topic in CYBERSECURITY_TOPICS:
        if re.search(rf"\b{topic}\b", user_input):
            return f"🔐 Cybersecurity:\n{CYBERSECURITY_TOPICS[topic]}"

    # -------------------------
    # EXAMS & ASSIGNMENTS
    # -------------------------
    if re.search(r"\b(exam|test|assignment|homework)\b", user_input):
        return "Please specify the subject and topic so I can assist effectively."

    # -------------------------
    # EXIT PATTERN
    # -------------------------
    # Uses sys.exit() to immediately and cleanly terminate the program.
    if re.search(r"\b(bye|exit|quit)\b", user_input):
        print(random.choice(GENERAL_RESPONSES["bye"]))
        sys.exit()

    # Default fallback response
    return random.choice(GENERAL_RESPONSES["default"])


# =========================
# MAIN PROGRAM
# =========================

def main():
    """
    Entry point of the chatbot application.
    Handles user interaction in a continuous loop.
    """
    print("\n🎓 EDUCATIONAL CHATBOT SYSTEM")
    print("Introduce yourself by saying: 'My name is ...'")
    print("Ask about topics, course codes, or exams.")
    print("Type 'bye' to exit.\n")

    while True:
        user_input = input("User: ")
        response = get_response(user_input)
        print("Chatbot:", response)


if __name__ == "__main__":
    main()
```

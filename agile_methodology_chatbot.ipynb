```python
import random
import sys

# AgileBot: A rule-based educational chatbot about Agile Methodology
# Built following your exact 10-step guide (see explanations below the code)

def get_response(user_input):
    text = user_input.lower().strip()

    # Greeting intent
    if any(greet in text for greet in ["hi", "hello", "hey", "greetings", "sup"]):
        return random.choice([
            "Hello! 👋 I'm AgileBot – your guide to everything Agile. What would you like to explore today?",
            "Hey there! Ready to master Agile Methodology? Ask me anything!",
            "Hi! Let's go agile together. What's on your mind?"
        ])

    # Help / capabilities intent
    elif any(word in text for word in ["help", "what can you", "capabilities", "topics"]):
        return ("I can teach you EVERYTHING about Agile:\n"
                "• What is Agile & the Manifesto\n"
                "• 12 Principles\n"
                "• Scrum (roles, events, artifacts)\n"
                "• Kanban\n"
                "• Benefits & Agile vs Waterfall\n"
                "• Examples, implementation tips, and even a quick quiz!\n"
                "Just ask naturally – e.g. 'Explain Scrum roles' or 'Give me the Agile Manifesto'.")

    # What is Agile intent
    elif any(phrase in text for phrase in ["what is agile", "agile methodology", "define agile", "agile overview"]):
        return ("Agile Methodology is an iterative and incremental approach to project management and software development. "
                "It focuses on flexibility, collaboration, customer satisfaction, and delivering working software quickly. "
                "Instead of rigid plans, Agile embraces change and continuous feedback.")

    # Agile Manifesto intent
    elif "manifesto" in text:
        return ("The Agile Manifesto (2001) states:\n\n"
                "We are uncovering better ways of developing software by doing it and helping others do it.\n"
                "Through this work we have come to value:\n"
                "1. Individuals and interactions over processes and tools\n"
                "2. Working software over comprehensive documentation\n"
                "3. Customer collaboration over contract negotiation\n"
                "4. Responding to change over following a plan\n\n"
                "That is, while there is value in the items on the right, we value the items on the left more.")

    # 12 Principles intent
    elif any(word in text for word in ["principles", "12 principles", "principle"]):
        return ("The 12 Principles behind the Agile Manifesto:\n\n"
                "1. Our highest priority is to satisfy the customer through early and continuous delivery of valuable software.\n"
                "2. Welcome changing requirements, even late in development.\n"
                "3. Deliver working software frequently (weeks rather than months).\n"
                "4. Business people and developers must work together daily.\n"
                "5. Build projects around motivated individuals. Give them the environment and support they need.\n"
                "6. The most efficient method of conveying information is face-to-face conversation.\n"
                "7. Working software is the primary measure of progress.\n"
                "8. Agile processes promote sustainable development. Maintain a constant pace indefinitely.\n"
                "9. Continuous attention to technical excellence and good design enhances agility.\n"
                "10. Simplicity – the art of maximizing the amount of work not done – is essential.\n"
                "11. The best architectures, requirements, and designs emerge from self-organizing teams.\n"
                "12. At regular intervals, the team reflects on how to become more effective.")

    # Scrum framework (with smart branching)
    elif "scrum" in text:
        if any(role in text for role in ["role", "owner", "master", "team", "product owner", "scrum master"]):
            return ("Scrum Roles:\n"
                    "• Product Owner – Maximizes product value, manages and prioritizes the Product Backlog.\n"
                    "• Scrum Master – Facilitates the Scrum process, removes impediments, coaches the team.\n"
                    "• Development Team – Cross-functional group that does the actual work (self-organizing).")
        elif any(event in text for event in ["event", "ceremony", "daily", "planning", "review", "retrospective", "sprint"]):
            return ("Scrum Events (all time-boxed):\n"
                    "• Sprint Planning (up to 8h) – Plan the work for the Sprint.\n"
                    "• Daily Scrum (15 min) – Stand-up sync: What did I do? What will I do? Any blockers?\n"
                    "• Sprint Review (4h max) – Demo the Increment and get feedback.\n"
                    "• Sprint Retrospective (3h max) – Inspect the process and improve.")
        elif any(artifact in text for artifact in ["artifact", "backlog", "increment", "product backlog", "sprint backlog"]):
            return ("Scrum Artifacts:\n"
                    "• Product Backlog – Ordered list of everything needed in the product.\n"
                    "• Sprint Backlog – Set of Product Backlog items selected for the Sprint + plan.\n"
                    "• Increment – Sum of all completed Product Backlog items at the end of a Sprint (must be 'Done').")
        else:
            return ("Scrum is the most popular Agile framework. It uses fixed-length iterations called Sprints (usually 1–4 weeks). "
                    "Ask me specifically about 'Scrum roles', 'Scrum events', or 'Scrum artifacts' for details!")

    # Kanban intent
    elif any(k in text for k in ["kanban", "board", "wip", "work in progress"]):
        return ("Kanban is a visual Agile method that uses a board with columns (To Do → In Progress → Done). "
                "Key practices:\n"
                "• Visualize the workflow\n"
                "• Limit Work In Progress (WIP limits) to avoid bottlenecks\n"
                "• Manage flow (pull system)\n"
                "• Make process policies explicit\n"
                "No fixed iterations – continuous delivery!")

    # Benefits of Agile intent
    elif any(b in text for b in ["benefit", "advantage", "why use agile", "why agile"]):
        return ("Key benefits of Agile:\n"
                "• Faster time-to-market\n"
                "• Higher customer satisfaction\n"
                "• Better adaptability to change\n"
                "• Improved team morale and collaboration\n"
                "• Reduced risk through early feedback\n"
                "• Higher product quality")

    # Agile vs Waterfall intent
    elif any(w in text for w in ["waterfall", "vs waterfall", "traditional", "compare"]):
        return ("Agile vs Waterfall:\n"
                "Waterfall = Linear, sequential, big upfront planning, changes are expensive.\n"
                "Agile = Iterative, flexible, welcomes change, delivers value early.\n"
                "Use Waterfall for fixed-scope, well-defined projects (e.g. construction).\n"
                "Use Agile for software, innovation, or any project where requirements evolve.")

    # Quick quiz intent (interactive style – one question at a time)
    elif "quiz" in text or "test me" in text or "question" in text:
        return ("Quick Agile Quiz!\n"
                "Q1: What does the Agile Manifesto value MORE than processes and tools?\n"
                "A) Documentation   B) Individuals and interactions\n"
                "Reply with your answer (A or B) and I'll tell you if you're right + explain!")

    # Example / implementation intent
    elif any(e in text for e in ["example", "implement", "how to start", "real world"]):
        return ("Real-world example: Spotify uses Scrum + Kanban (Squads, Tribes, Chapters). "
                "They deliver features every week instead of every 6 months. "
                "Want a step-by-step 'How to adopt Agile in your team' guide?")

    # Farewell intent
    elif any(fare in text for fare in ["bye", "goodbye", "quit", "exit", "thank", "thanks", "see you"]):
        return random.choice([
            "You're welcome! Keep being agile 🚀",
            "Glad I could help. Stay iterative!",
            "Bye! Come back anytime for more Agile wisdom."
        ])

    # Fallback (never dead-end)
    else:
        return ("Hmm, I didn't catch that one. Try asking about:\n"
                "• The Agile Manifesto\n"
                "• Scrum roles/events\n"
                "• Kanban board\n"
                "• Benefits or Agile vs Waterfall\n"
                "• Or just say 'quiz' or 'help'!\n"
                "Rephrase and I'll get it right.")

# Main chatbot loop
def run_agile_bot():
    print("=" * 60)
    print("          🤖 AGILEBOT – Master Agile Methodology")
    print("Type 'quit', 'exit', or 'bye' to stop. Ask anything!")
    print("=" * 60)

    while True:
        try:
            user_input = input("\nYou: ").strip()
            if user_input.lower() in ["quit", "exit", "bye"]:
                print("AgileBot: Goodbye! Remember – inspect & adapt! 👋")
                break
            response = get_response(user_input)
            print(f"AgileBot: {response}")
        except KeyboardInterrupt:
            print("\n\nAgileBot: Session ended. Stay agile!")
            break

if __name__ == "__main__":
    run_agile_bot()
```

```python
# =====================================================
# AGILEBOT – Now with More Agile Frameworks
# Running directly in Google Colab (pure Python)
# Features: Greetings, 12 Principles, Quiz with memory,
#           Context follow-ups + NEW: XP, SAFe, FDD, Lean, Crystal
# =====================================================

import random
import re

# In-memory conversation context
user_context = {
    'quiz_active': False,
    'last_topic': None
}

def get_response(user_input):
    text = user_input.lower().strip()
    ctx = user_context

    # ==================== GREETINGS ====================
    greetings = ["hi", "hello", "hey", "greetings", "sup", "good morning",
                 "good afternoon", "good evening", "howdy", "yo", "hola"]

    if any(greet in text for greet in greetings):
        ctx['last_topic'] = None
        ctx['quiz_active'] = False
        return random.choice([
            "Hello! 👋 Welcome to AgileBot! Ready to explore even more Agile frameworks?",
            "Hey there! 🚀 Let's dive deeper into Agile. What would you like to learn today?",
            "Hi! Glad you're here. Ask about Scrum, Kanban, XP, SAFe, or any principle!"
        ])

    # ==================== INTERACTIVE QUIZ ====================
    if ctx['quiz_active']:
        if text in ["a", "b", "a)", "b)"]:
            ctx['quiz_active'] = False
            if "a" in text:
                return ("❌ Incorrect. The Agile Manifesto values **Individuals and interactions** over processes and tools.\n"
                        "Type 'quiz' to try again!")
            else:
                return ("✅ Correct! Great job.\nType 'quiz' for more or ask about frameworks like SAFe or XP.")

    if "quiz" in text:
        ctx['quiz_active'] = True
        return ("🧠 Quick Quiz!\n"
                "Q: What does the Agile Manifesto value MORE than processes and tools?\n"
                "A) Documentation   B) Individuals and interactions\n\n"
                "Reply with A or B!")

    # ==================== 12 PRINCIPLES ====================
    if any(p in text for p in ["12 principles", "principles", "principle list", "full principles"]):
        ctx['last_topic'] = "principles"
        return ("📜 **The 12 Principles of the Agile Manifesto**:\n\n"
                "1. Satisfy the customer through early & continuous delivery...\n"
                "2. Welcome changing requirements...\n"
                "... (full list shown earlier)\n\n"
                "Type 'principle 5' (or any number) for details!")

    if "principle" in text:
        match = re.search(r'principle\s*(\d+)', text)
        if match:
            num = int(match.group(1))
            if 1 <= num <= 12:
                ctx['last_topic'] = "principles"
                details = {
                    1: "Focuses on delivering real value to the customer as quickly as possible.",
                    2: "Change is not the enemy — adapt even late in development.",
                    3: "Short iterations preferred over long ones.",
                    4: "Daily collaboration between business and developers.",
                    5: "Build around motivated individuals and trust them.",
                    6: "Face-to-face conversation is most efficient.",
                    7: "Working software is the primary measure of progress.",
                    8: "Maintain a sustainable constant pace.",
                    9: "Continuous attention to technical excellence.",
                    10: "Simplicity — maximize work not done.",
                    11: "Self-organizing teams produce best designs.",
                    12: "Regular reflection and adjustment (Retrospectives)."
                }
                return f"**Principle #{num}**: {details.get(num)}\n\nAsk for another or '12 principles'."
            else:
                return "Please ask for principle 1–12."
        else:
            return "Type 'principle 5' or '12 principles'."

    # ==================== NEW AGILE FRAMEWORKS ====================
    if any(x in text for x in ["xp", "extreme programming", "extreme"]):
        ctx['last_topic'] = "xp"
        return ("**Extreme Programming (XP)**:\n"
                "• Focuses on technical excellence and high-quality code.\n"
                "• Key practices: Pair Programming, Test-Driven Development (TDD), Continuous Integration, Frequent Releases.\n"
                "• Values: Communication, Simplicity, Feedback, Courage, Respect.\n"
                "Best for: Teams that want strong engineering discipline.")

    elif any(s in text for s in ["safe", "scaled agile framework", "scaled agile"]):
        ctx['last_topic'] = "safe"
        return ("**Scaled Agile Framework (SAFe)**:\n"
                "• Designed for large enterprises and multiple teams.\n"
                "• Combines Agile, Lean, and DevOps practices.\n"
                "• Levels: Team, Program (ART), Large Solution, Portfolio.\n"
                "• Includes roles like Release Train Engineer (RTE) and PI Planning.\n"
                "Best for: Organizations scaling Agile across 50+ people.")

    elif any(f in text for f in ["fdd", "feature driven development", "feature-driven"]):
        ctx['last_topic'] = "fdd"
        return ("**Feature-Driven Development (FDD)**:\n"
                "• Model-driven and feature-centric approach.\n"
                "• Process: Develop overall model → Build feature list → Plan by feature → Design & build by feature.\n"
                "• Short iterations focused on delivering client-valued features.\n"
                "Best for: Larger projects needing clear feature tracking.")

    elif any(l in text for l in ["lean", "lean software development"]):
        ctx['last_topic'] = "lean"
        return ("**Lean Software Development**:\n"
                "• Inspired by Lean manufacturing (Toyota).\n"
                "• Core principles: Eliminate waste, Amplify learning, Decide late, Deliver fast, Empower team, Build integrity, See the whole.\n"
                "• Focuses on maximizing customer value while minimizing waste.\n"
                "Often combined with Kanban.")

    elif any(c in text for c in ["crystal", "crystal methodology"]):
        ctx['last_topic'] = "crystal"
        return ("**Crystal**:\n"
                "• Family of lightweight, human-powered methods (Crystal Clear, Orange, Red, etc.).\n"
                "• Tailored to team size and project criticality.\n"
                "• Emphasizes people, frequent delivery, and reflective improvement.\n"
                "• Less prescriptive — focuses on communication and safety.")

    # ==================== EXISTING FRAMEWORKS (kept & improved) ====================
    elif "scrum" in text:
        ctx['last_topic'] = "scrum"
        if any(r in text for r in ["role", "owner", "master", "team"]):
            return ("**Scrum Roles**:\n• Product Owner\n• Scrum Master\n• Development Team")
        else:
            return "Scrum uses fixed-length Sprints. Ask 'scrum roles' or 'scrum events' for details!"

    elif any(k in text for k in ["kanban", "board", "wip"]):
        ctx['last_topic'] = "kanban"
        return ("**Kanban**: Visual board, WIP limits, continuous flow. No fixed sprints.")

    elif any(b in text for b in ["benefit", "advantage", "why agile"]):
        ctx['last_topic'] = "benefits"
        return ("Benefits: Faster delivery, higher quality, better adaptability, happier teams.")

    elif any(w in text for w in ["waterfall", "vs"]):
        ctx['last_topic'] = "comparison"
        return ("Agile = flexible & iterative. Waterfall = linear & rigid.")

    # Context-aware follow-up
    elif any(f in text for f in ["more", "tell me more", "details", "explain", "elaborate", "compare"]):
        if ctx['last_topic']:
            topic = ctx['last_topic'].upper()
            return f"Want more on **{topic}**? Ask specifically (e.g. 'XP practices' or 'SAFe levels')."
        else:
            return "What would you like more details on? Try 'SAFe', 'XP', '12 principles', or 'quiz'."

    # Farewell
    elif any(fare in text for fare in ["bye", "goodbye", "see you", "exit", "quit", "thanks", "thank you"]):
        ctx['quiz_active'] = False
        ctx['last_topic'] = None
        return random.choice([
            "You're welcome! Keep inspecting & adapting 🚀",
            "Glad I could help with all these frameworks. Stay agile!",
            "Bye! Come back anytime."
        ])

    # Fallback
    else:
        return ("I didn't catch that. Try:\n"
                "• 'hi'\n"
                "• '12 principles' or 'principle 5'\n"
                "• 'quiz'\n"
                "• 'What is XP?' or 'SAFe' or 'Kanban'\n"
                "• 'FDD', 'Lean', or 'Crystal'\n"
                "• 'tell me more'")

# ==================== RUN IN COLAB ====================
print("=" * 75)
print("          AGILEBOT")
print("          Scrum • Kanban • XP • SAFe • FDD • Lean • Crystal")
print("Type 'quit', 'exit', or 'bye' to stop.")
print("=" * 75)

while True:
    try:
        user_input = input("\nYou: ").strip()
        if user_input.lower() in ["quit", "exit", "bye"]:
            print("AgileBot: Goodbye! Remember – inspect & adapt! 👋")
            break
        response = get_response(user_input)
        print(f"AgileBot: {response}")
    except KeyboardInterrupt:
        print("\n\nAgileBot: Session ended. Stay agile!")
        break
```
